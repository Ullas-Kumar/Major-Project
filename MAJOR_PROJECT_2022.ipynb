# TRAIN TEST SPLIT



```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
X_train = np.array(X_train)
```

```python
Y_train = np.array(Y_train)
```

```python
X_test = np.array(X_test)
```

```python
Y_test = np.array(Y_test)
```

```python
len(Y_test)
```

# MODEL 1 - LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
predictions_model1 = lr.predict(X_test)
predictions_model1
```

```python
for i,j in enumerate(predictions_model1):
    if(j<0.5):
        predictions_model1[i]=0
    else:
        predictions_model1[i]=1
print(predictions_model1)
predictions_model1.shape
```

```python
Y_test = np.array(Y_test)
```

```python
print(Y_test)
Y_test.shape
```

```python
print("X_test_array     Y_test")
for i,j in zip(predictions_model1, Y_test):
    print(i, "              ", j)
```

```python
print("Linear Regression Accuracy :\n")
lr.score(X_test, Y_test)
```

```python
Y_pred= lr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.plot(X_test, Y_pred, color='blue')
```

```python
from sklearn.metrics import confusion_matrix
```

```python
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(Y_test, predictions_model1)
import seaborn as sns
plt.figure(figsize = (8,6))
fg = sns.heatmap(cm,annot=True,cmap='Oranges')
figure = fg.get_figure()
plt.xlabel("Predicted")
plt.ylabel("Actual")
```

```python
tp, fn, fp, tn = cm.ravel()

precision = tp / ( tp + fp )
recall =  tp / ( tp + fn )
fiscore=2*(precision*recall)/(precision+recall)

print('precision -', precision)
print('recall -', recall)
print('fi_score -',fiscore)
```

# MODEL 7 - NEURAL NETWORK


```python
import tensorflow
```

```python
import keras
import matplotlib
from matplotlib import pyplot as plt
```

```python
from keras.layers import Dense 
from keras.layers import Flatten
from keras.utils.vis_utils import plot_model
```

```python
from keras.backend import flatten
model7=keras.Sequential()

model7.add(Dense(1,input_dim=753,activation='sigmoid'))
model7.add(Dense(400,activation='sigmoid'))
model7.add(Dense(300,activation='sigmoid'))
model7.add(Dense(200,activation='sigmoid'))
model7.add(Dense(100,activation='sigmoid'))
model7.add(Dense(1,activation='sigmoid'))

model7.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model7.fit(X_train,Y_train,epochs=1000)
```

```python
model7.evaluate(X_test,Y_test)
```

```python
prednn=model7.predict(X_test)
prednn
```

```python
for i,j in enumerate(prednn):
  if(j<0.5):
      prednn[i]=0
  else:
      prednn[i]=1
```

```python
prednn.reshape(1,152)
predd= prednn.flatten()
print(predd)

```

```python
Y_test
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
cm=confusion_matrix(Y_test,predd)
plt.figure(figsize=(8,6))
plt.title('Neural Network')
fg=sn.heatmap(cm,annot=True)
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
